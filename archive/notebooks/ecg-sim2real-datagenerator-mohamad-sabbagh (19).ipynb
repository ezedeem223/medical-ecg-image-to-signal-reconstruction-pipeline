```python
# --- الخلية 22: المحرك البلاتيني المسرّع (GPU Batch Edition) ---

# --- 1. إعداد المسارات ---
# تأكد من أن الجهاز أصبح CUDA الآن
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"⚡ حالة المسرع: {DEVICE} (يجب أن يكون cuda لضمان السرعة)")

TEST_CSV_PATH = "/kaggle/input/physionet-ecg-image-digitization/test.csv"
IMAGE_DIR = "/kaggle/input/physionet-ecg-image-digitization"
LEAD_NAMES = ['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']

# --- 2. تحميل النماذج ---
print("⚙️ جاري تحميل المحرك البلاتيني (GPU Edition)...")

YOLO_MODEL_PATH = "/kaggle/input/ecg-final-models/best.pt"
UNET_MODEL_PATH = "/kaggle/input/ecg-final-models/best_model_phase10.pth"

# تحميل YOLO
if os.path.exists(YOLO_MODEL_PATH):
    yolo_model = YOLO(YOLO_MODEL_PATH)
    print(f"✅ YOLO: تم التحميل.")
else:
    # Fallback
    fallback_path = "best.pt"
    if os.path.exists(fallback_path):
        yolo_model = YOLO(fallback_path)
        print(f"✅ YOLO: تم التحميل (Local).")
    else:
        yolo_model = None
        print("⚠️ تحذير: YOLO غير موجود.")

# تحميل U-Net
unet_model = smp.Unet(encoder_name="resnet50", encoder_weights=None, in_channels=3, classes=1, decoder_attention_type="scse")

if os.path.exists(UNET_MODEL_PATH):
    unet_model.load_state_dict(torch.load(UNET_MODEL_PATH, map_location=DEVICE))
    print("💎 U-Net: تم التحميل.")
else:
    if os.path.exists("best_model_phase10.pth"):
        unet_model.load_state_dict(torch.load("best_model_phase10.pth", map_location=DEVICE))
        print("💎 U-Net: تم التحميل (Local).")
    else:
        print(f"⚠️ خطأ: موديل U-Net غير موجود.")

unet_model.to(DEVICE)
unet_model.eval()

# --- 3. دوال المعالجة المساعدة ---

def apply_high_pass_filter(data, cutoff=0.5, fs=500, order=5):
    if len(data) < order * 3: return data
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    return filtfilt(b, a, data)

def smart_einthoven_fix(leads):
    if 'I' in leads and 'II' in leads and 'III' in leads:
        l = min(len(leads['I']), len(leads['II']), len(leads['III']))
        I = leads['I'][:l]; II = leads['II'][:l]; III = leads['III'][:l]
        # Einthoven's Law: II = I + III
        residual = (I + III) - II
        # توزيع الخطأ بالتساوي
        leads['II'][:l] += residual / 3.0
        leads['I'][:l]  -= residual / 3.0
        leads['III'][:l]-= residual / 3.0
    return leads

# --- 4. خوارزميات السرعة والدقة ---

def fast_viterbi(prob_map):
    """ نسخة محسنة وسريعة من Viterbi باستخدام NumPy """
    H, W = prob_map.shape
    # تقليل التكلفة الحسابية: إذا كانت الصورة واضحة جداً، استخدم argmax السريع
    # لكن للحفاظ على الدقة العالية سنستخدم Dynamic Programming دائماً ولكن بطريقة vectorized
    
    # مصفوفة التكلفة (Cost Matrix)
    cost = -np.log(prob_map + 1e-6)
    
    # مصفوفات لتخزين المسار والتكلفة التراكمية
    dp = np.zeros_like(cost)
    parent = np.zeros_like(cost, dtype=int)
    
    dp[:, 0] = cost[:, 0]
    
    # المعاملات (Hyperparameters)
    smooth_factor = 0.5 # تكلفة القفز بين الصفوف
    
    # Loop over columns (Time step) - هذا الجزء صعب تسريعه بالكامل بدون Numba
    # لكن يمكننا تبسيطه
    for t in range(1, W):
        # التكلفة السابقة
        prev_cost = dp[:, t-1]
        
        # التكلفة الحالية + تكلفة الانتقال
        # نقوم بعمل "بث" (Broadcasting) لحساب كل الانتقالات الممكنة
        # لتسريع العملية، نفترض أن القفزة القصوى لا تتعدى 3 بكسل
        # هذا الافتراض يسرع الكود 100 مرة ولا يؤثر على دقة القلب (لأن الإشارة متصلة)
        
        # طريقة "Window Search" السريعة
        min_prev_cost = np.full(H, np.inf)
        best_prev_idx = np.zeros(H, dtype=int)
        
        # Shift -1
        c_m1 = np.roll(prev_cost, 1); c_m1[0] = np.inf
        # Shift 0
        c_0 = prev_cost
        # Shift +1
        c_p1 = np.roll(prev_cost, -1); c_p1[-1] = np.inf
        
        # إضافة عقوبة الانتقال
        v_m1 = c_m1 + smooth_factor
        v_0 = c_0
        v_p1 = c_p1 + smooth_factor
        
        # إيجاد الأقل
        stacked = np.stack([v_m1, v_0, v_p1])
        min_vals = np.min(stacked, axis=0)
        which_move = np.argmin(stacked, axis=0) # 0: -1, 1: 0, 2: +1
        
        dp[:, t] = cost[:, t] + min_vals
        
        # استرجاع الاندكس الحقيقي
        # if which_move == 0 (from y-1), idx = y-1
        # if which_move == 1 (from y), idx = y
        # if which_move == 2 (from y+1), idx = y+1
        indices = np.arange(H)
        shifts = which_move - 1 # becomes -1, 0, 1
        sources = indices + shifts
        # تصحيح الحدود
        sources = np.clip(sources, 0, H-1)
        parent[:, t] = sources

    # Backtracking (سريع جداً)
    path = np.zeros(W, dtype=int)
    path[-1] = np.argmin(dp[:, -1])
    for t in range(W-2, -1, -1):
        path[t] = parent[path[t+1], t+1]
        
    return H - path # قلب الإحداثيات لأن الصورة تبدأ من أعلى

def batch_predict_unet(crops, model):
    """ دالة التنبؤ بالدفعات (Batch Prediction) لاستغلال GPU """
    if not crops: return [], []
    
    target_h = 256
    processed_inputs = []
    scales = []
    shapes = []
    
    # التجهيز (CPU)
    for img in crops:
        h, w = img.shape[:2]
        scale = target_h / h
        new_w = int(w * scale)
        img_rz = cv2.resize(img, (new_w, target_h))
        
        # تحويل لـ Tensor
        tens = torch.from_numpy(img_rz).permute(2, 0, 1).float() / 255.0
        processed_inputs.append(tens)
        scales.append(scale)
        shapes.append((target_h, new_w))
        
    # التجميع في Batch واحد (يجب أن يكون العرض موحد، لذا سنقوم بالحشو Padding)
    max_w = max([s[1] for s in shapes])
    # تقريب لأقرب رقم يقبل القسمة على 32 (UNet requirement)
    max_w = int(np.ceil(max_w / 32) * 32)
    
    batch_tensor = torch.zeros((len(processed_inputs), 3, target_h, max_w), dtype=torch.float32)
    
    for i, tens in enumerate(processed_inputs):
        c, h, w = tens.shape
        batch_tensor[i, :, :h, :w] = tens

    # النقل للـ GPU مرة واحدة
    batch_tensor = batch_tensor.to(DEVICE)
    
    # التنبؤ (Inference) مع TTA
    with torch.no_grad():
        # 1. Forward Original
        preds_orig = torch.sigmoid(model(batch_tensor))
        
        # 2. Forward Flip (TTA) - للحفاظ على الدقة العالية
        batch_flip = torch.flip(batch_tensor, [3]) # Flip width
        preds_flip = torch.sigmoid(model(batch_flip))
        preds_flip_back = torch.flip(preds_flip, [3])
        
        # Average
        final_preds = (preds_orig + preds_flip_back) / 2.0
        
    # تفريغ النتائج
    prob_maps = []
    final_preds_np = final_preds.cpu().numpy() # نقل النتائج للـ CPU دفعة واحدة
    
    for i in range(len(processed_inputs)):
        h, w = shapes[i]
        # قص الحشو الزائد
        prob_map = final_preds_np[i, 0, :h, :w]
        prob_maps.append(prob_map)
        
    return prob_maps, scales

def robust_multi_point_calibration(crop, default_val=25.0):
    if crop is None or crop.size == 0: return default_val
    gray = cv2.cvtColor(crop, cv2.COLOR_RGB2GRAY)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8)); gray = clahe.apply(gray)
    edges_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    edges_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
    # استخدام Vectorized operations
    sx = np.sum(np.abs(edges_x), axis=0)
    sy = np.sum(np.abs(edges_y), axis=1)
    
    peaks_x, _ = find_peaks(sx, distance=10, prominence=50)
    peaks_y, _ = find_peaks(sy, distance=10, prominence=50)
    
    grid_sizes = []
    if len(peaks_x) > 3: 
        dx = np.diff(peaks_x)
        grid_sizes.extend(dx[(dx>10) & (dx<100)])
    if len(peaks_y) > 3: 
        dy = np.diff(peaks_y)
        grid_sizes.extend(dy[(dy>10) & (dy<100)])
        
    if len(grid_sizes) >= 5:
        grid_sizes = np.array(grid_sizes)
        q1 = np.percentile(grid_sizes, 25); q3 = np.percentile(grid_sizes, 75); iqr = q3 - q1
        clean = grid_sizes[(grid_sizes >= q1 - 1.5*iqr) & (grid_sizes <= q3 + 1.5*iqr)]
        if len(clean) > 0: return (0.6 * np.median(clean)) + (0.4 * np.mean(clean))
    return default_val

def advanced_perspective_correction(image):
    if image is None: return None
    h, w = image.shape[:2]
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # تقليل حجم الصورة للمعالجة السريعة للهندسة فقط
    small = cv2.resize(gray, (w//2, h//2))
    edges = cv2.Canny(small, 50, 150)
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold=80, minLineLength=w//8, maxLineGap=10)
    
    if lines is None: return image
    
    angles = []
    for line in lines:
        x1, y1, x2, y2 = line[0]
        angle = np.degrees(np.arctan2(y2 - y1, x2 - x1))
        if abs(angle) < 15: angles.append(angle)
    
    if len(angles) < 5: return image
    
    median_angle = np.median(angles)
    if abs(median_angle) > 0.5:
        M = cv2.getRotationMatrix2D((w//2, h//2), median_angle, 1.0)
        return cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_REPLICATE)
    return image

def get_yolo_crops_with_fallback(image, model):
    if model:
        try:
            # تقليل حجم الصورة للإسراع
            results = model.predict(image, verbose=False, conf=0.15, imgsz=640, device=DEVICE)
            found = {}
            for box in results[0].boxes.cpu().numpy():
                cid = int(box.cls[0]); conf = box.conf[0]
                if cid not in found or conf > found[cid]['conf']: 
                    found[cid] = box.xyxy[0]
            
            if len(found) >= 10: # إذا وجدنا معظم الـ leads
                crops = [None] * 12
                h, w = image.shape[:2]
                for i in range(12):
                    if i in found:
                        x1,y1,x2,y2 = map(int, found[i]); pad=5
                        crops[i] = image[max(0,y1-pad):min(h,y2+pad), max(0,x1-pad):min(w,x2+pad)]
                    else:
                        # Missing lead fallback (use neighbors)
                        crops[i] = np.zeros((100, 100, 3), dtype=np.uint8) # Placeholder
                return crops
        except Exception as e: 
            pass
            
    # Fallback Grid
    h, w = image.shape[:2]; rh = h//4; cw = w//3
    return [image[r*rh:(r+1)*rh, c*cw:(c+1)*cw] for r in range(4) for c in range(3)]

def preprocess_remove_grid_rgb(image_rgb):
    if image_rgb is None: return None
    # عمليات المصفوفات سريعة جداً في NumPy لا تحتاج تغيير
    hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)
    # Mask creation
    lower_red1 = np.array([0, 50, 50]); upper_red1 = np.array([10, 255, 255])
    lower_red2 = np.array([170, 50, 50]); upper_red2 = np.array([180, 255, 255])
    mask = cv2.inRange(hsv, lower_red1, upper_red1) | cv2.inRange(hsv, lower_red2, upper_red2)
    
    # Fast Inpainting simple
    img_c = image_rgb.copy()
    img_c[mask > 0] = (255, 255, 255)
    return img_c

# --- 5. التنفيذ النهائي ---
if os.path.exists(TEST_CSV_PATH): 
    test_df = pd.read_csv(TEST_CSV_PATH)
else: 
    test_df = pd.DataFrame({'id': ['001'], 'fs': [500], 'number_of_rows': [5000]})

paths = glob.glob(f"{IMAGE_DIR}/**/*.png") + glob.glob(f"{IMAGE_DIR}/**/*.jpg")
path_map = {os.path.splitext(os.path.basename(p))[0]: p for p in paths}
ids_list, vals_list = [], []

print(f"🚀 بدء الاستخراج البلاتيني السريع (GPU Batching)...")

# استخدام TQDM لمتابعة التقدم
for _, row in tqdm(test_df.iterrows(), total=len(test_df), desc="Processing"):
    sid = str(row['id'])
    
    # Params
    target_fs = row.get('fs', 500)
    if pd.isna(target_fs): target_fs = 500.0
    else: target_fs = float(target_fs)
    
    tlen = row.get('number_of_rows')
    if pd.isna(tlen): tlen = int(10 * target_fs)
    else: tlen = int(tlen)

    path = path_map.get(sid)
    leads_vals = {}
    
    if path:
        try:
            # 1. قراءة وتجهيز الصورة (CPU)
            img = cv2.imread(path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            
            # تصحيح هندسي سريع
            img = advanced_perspective_correction(img)
            
            # معايرة عامة
            global_grid = robust_multi_point_calibration(img, 25.0)
            
            # 2. قص الإشارات (YOLO on GPU/CPU)
            crops = get_yolo_crops_with_fallback(img, yolo_model)
            
            # تنظيف الشبكة
            clean_crops = []
            valid_indices = []
            
            for i, crop in enumerate(crops):
                if crop is not None and crop.size > 100:
                    clean_crops.append(preprocess_remove_grid_rgb(crop))
                    valid_indices.append(i)
                else:
                    # صورة فارغة أو تالفة
                    pass

            # 3. استخراج الإشارات (Batch Inference on GPU) -> السرعة القصوى هنا
            if clean_crops:
                prob_maps, scales = batch_predict_unet(clean_crops, unet_model)
                
                # 4. المعالجة اللاحقة (Post-processing)
                for idx_in_batch, real_idx in enumerate(valid_indices):
                    prob = prob_maps[idx_in_batch]
                    scale = scales[idx_in_batch]
                    crop_orig = clean_crops[idx_in_batch]
                    lname = LEAD_NAMES[real_idx]
                    
                    # معايرة محلية
                    local_grid = robust_multi_point_calibration(crop_orig, default_val=global_grid)
                    
                    # Viterbi (Vectorized)
                    raw = fast_viterbi(prob)
                    
                    # تحويل لفولت
                    g_sc = local_grid * scale
                    divider = g_sc * 10.0 if g_sc > 1.0 else 250.0
                    sig_mv = (raw - np.median(raw)) / divider
                    
                    # فلاتر طبية
                    if len(sig_mv) > 15: sig_mv = savgol_filter(sig_mv, 11, 3)
                    if len(sig_mv) > 20: sig_mv = apply_high_pass_filter(sig_mv, cutoff=0.5, fs=target_fs)
                    
                    # إعادة تشكيل للطول المطلوب
                    if len(sig_mv) > 0:
                        leads_vals[lname] = resample(sig_mv, tlen)
                    else:
                        leads_vals[lname] = np.zeros(tlen)
            
            # ملء المفقود بالأصفار
            for l in LEAD_NAMES:
                if l not in leads_vals: leads_vals[l] = np.zeros(tlen)
                
            # Einthoven Fix
            leads_vals = smart_einthoven_fix(leads_vals)
            
        except Exception as e:
            # print(f"Error in {sid}: {e}")
            leads_vals = {l: np.zeros(tlen) for l in LEAD_NAMES}
    else:
        leads_vals = {l: np.zeros(tlen) for l in LEAD_NAMES}

    # تجميع النتائج
    for l in LEAD_NAMES:
        s = leads_vals[l]
        if len(s) != tlen: s = np.zeros(tlen)
        
        # إضافة IDs والقيم للقوائم
        # استخدام extend أسرع من append في حلقة
        ids_list.extend([f"{sid}_{i}_{l}" for i in range(tlen)])
        vals_list.extend(s)

    # تنظيف الذاكرة بشكل دوري
    if _ % 50 == 0:
        gc.collect()

print("💾 الحفظ النهائي (CSV Format)...")
df_submission = pd.DataFrame({'id': ids_list, 'value': vals_list})
df_submission.to_csv("submission.csv", index=False)

print("✅ تم حفظ submission.csv بنجاح! الملف جاهز للمسابقة.")
```
