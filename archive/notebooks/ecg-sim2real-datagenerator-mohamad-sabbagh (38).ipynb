```python
# --- Cell 22: V46 BALANCED (Template Lengths + Light Sniper Gate + TTA) ---
import os, gc, csv, glob
import numpy as np
import pandas as pd
import cv2
import torch
from collections import OrderedDict
from tqdm import tqdm
from scipy.signal import savgol_filter, resample, find_peaks, butter, filtfilt
import segmentation_models_pytorch as smp
from ultralytics import YOLO

# =========================
# 0) CONFIG
# =========================
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

DATA_DIR = "/kaggle/input/physionet-ecg-image-digitization"
SAMPLE_PARQUET_PATH = f"{DATA_DIR}/sample_submission.parquet"
SUBMISSION_FILE = "submission.csv"

YOLO_PATH = "/kaggle/input/ecg-final-models/best.pt"
UNET_PATH = "/kaggle/input/ecg-final-models/best_model_phase10.pth"

LEAD_NAMES = ['I','II','III','aVR','aVL','aVF','V1','V2','V3','V4','V5','V6']
LEAD_TO_IDX = {n:i for i,n in enumerate(LEAD_NAMES)}

TARGET_H = 256
MAX_W = 2048
YOLO_CONF = 0.12
MAX_CACHE = 12
DEFAULT_FS = 500.0

# Light quality thresholds (NOT aggressive)
P2P_MIN = 0.03   # mV (too flat -> likely wrong calibration/seg)
P2P_MAX = 6.0    # mV (too huge -> likely wrong calibration)
ROUGH_MAX = 0.55 # loose noise gate

print(f"⚡ Device: {DEVICE}")

# =========================
# 1) TEMPLATE IDS + LENGTHS (SAFE SOURCE)
# =========================
if not os.path.exists(SAMPLE_PARQUET_PATH):
    raise FileNotFoundError("❌ sample_submission.parquet not found")

print("📦 Reading template ids...")
template = pd.read_parquet(SAMPLE_PARQUET_PATH, columns=["id"])
template_ids = template["id"].astype(str).to_numpy()
del template
gc.collect()

print("🧠 Mapping patient lengths from template (SAFE)...")
patient_len = {}
for rid in template_ids:
    try:
        pid, idx, lead = rid.rsplit("_", 2)
        idx = int(idx)
        cur = patient_len.get(pid, 0)
        if idx + 1 > cur:
            patient_len[pid] = idx + 1
    except:
        pass

print(f"✅ Lengths mapped for {len(patient_len)} pids.")

# =========================
# 2) INDEX IMAGES
# =========================
print("🗂️ Indexing images...")
image_paths = glob.glob(f"{DATA_DIR}/**/*.png", recursive=True) + glob.glob(f"{DATA_DIR}/**/*.jpg", recursive=True)
path_map = {os.path.splitext(os.path.basename(p))[0]: p for p in image_paths}
print(f"✅ Indexed images: {len(path_map):,}")

# =========================
# 3) LOAD MODELS + LEAD MAP
# =========================
print("⚙️ Loading models...")
yolo_model = None
CLASS_TO_LEAD_IDX = {i:i for i in range(12)}  # fallback identity

if os.path.exists(YOLO_PATH):
    try:
        yolo_model = YOLO(YOLO_PATH)
        tmp = {}
        if hasattr(yolo_model, "names"):
            for cid, cname in yolo_model.names.items():
                n = str(cname).strip().replace("Lead","").replace("lead","").replace(" ","")
                n = n.replace("AVR","aVR").replace("AVL","aVL").replace("AVF","aVF")
                if n in LEAD_TO_IDX:
                    tmp[int(cid)] = LEAD_TO_IDX[n]
        for i in range(12):
            tmp.setdefault(i, i)
        CLASS_TO_LEAD_IDX = tmp
        print(f"✅ YOLO loaded. CLASS_TO_LEAD_IDX: {CLASS_TO_LEAD_IDX}")
    except:
        yolo_model = None

unet_model = smp.Unet(
    encoder_name="resnet50",
    encoder_weights=None,
    in_channels=3,
    classes=1,
    decoder_attention_type="scse"
)
if os.path.exists(UNET_PATH):
    unet_model.load_state_dict(torch.load(UNET_PATH, map_location=DEVICE))
unet_model.to(DEVICE).eval()
print("✅ UNet ready.")

# =========================
# 4) HELPERS
# =========================
def clean_pid(pid):
    s = str(pid).strip()
    return s[:-2] if s.endswith(".0") else s

def get_image_path(pid):
    pid_s = clean_pid(pid)
    p = path_map.get(pid_s)
    if not p:
        try:
            p = path_map.get(str(int(float(pid_s))))
        except:
            pass
    return p

def safe_read_rgb(path):
    img = cv2.imread(path)
    if img is None:
        return None
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def preprocess_remove_red_grid(img):
    # remove red-ish pixels (safe)
    try:
        hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
        mask = cv2.inRange(hsv, (0,50,50), (10,255,255)) | cv2.inRange(hsv, (170,50,50), (180,255,255))
        out = img.copy()
        out[mask > 0] = 255
        return out
    except:
        return img

def robust_grid_spacing_px(img):
    # estimate spacing between grid lines (rough), fallback None if bad
    try:
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        edges = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)  # horizontal lines
        s = np.sum(np.abs(edges), axis=1)
        peaks, _ = find_peaks(s, distance=10, prominence=np.percentile(s, 85))
        if len(peaks) > 4:
            diffs = np.diff(peaks)
            valid = diffs[(diffs > 6) & (diffs < 200)]
            if len(valid) > 2:
                v = float(np.median(valid))
                if 6 <= v <= 200:
                    return v
    except:
        pass
    return None

def apply_filters(x, fs):
    try:
        x = np.nan_to_num(x)
        nyq = 0.5 * fs
        if nyq <= 0:
            return x
        b, a = butter(3, 0.5/nyq, btype="high")
        return filtfilt(b, a, x)
    except:
        return x

def lead_quality_ok(sig_mv):
    # Light sniper gate: only reject obvious failures
    if sig_mv is None or len(sig_mv) < 50:
        return False
    if not np.all(np.isfinite(sig_mv)):
        return False

    p95 = float(np.percentile(sig_mv, 95))
    p05 = float(np.percentile(sig_mv, 5))
    p2p = p95 - p05

    if p2p < P2P_MIN or p2p > P2P_MAX:
        return False

    d2 = np.diff(sig_mv, 2)
    rough = float(np.mean(np.abs(d2))) / (p2p + 1e-6)
    if rough > ROUGH_MAX:
        return False

    return True

def get_crops(img):
    h, w = img.shape[:2]
    crops = [None]*12

    # fallback grid crops (always available)
    rh, cw = h//4, w//3
    grid = [img[r*rh:(r+1)*rh, c*cw:(c+1)*cw] for r in range(4) for c in range(3)]

    det_count = 0
    if yolo_model is not None:
        try:
            scale = 1280 / max(h, w)
            if scale < 1:
                img_inf = cv2.resize(img, (int(w*scale), int(h*scale)))
            else:
                img_inf = img
                scale = 1.0

            res = yolo_model.predict(img_inf, conf=YOLO_CONF, verbose=False, device=DEVICE)
            if res and res[0].boxes is not None:
                for box in res[0].boxes.data.detach().cpu().numpy():
                    cls = int(box[5])
                    tgt = CLASS_TO_LEAD_IDX.get(cls, cls)
                    if 0 <= tgt < 12:
                        x1, y1, x2, y2 = box[:4] / scale
                        x1, y1 = max(0,int(x1)), max(0,int(y1))
                        x2, y2 = min(w,int(x2)), min(h,int(y2))
                        if (x2-x1) > 10 and (y2-y1) > 10:
                            crops[tgt] = img[y1:y2, x1:x2]
        except:
            pass

    for i in range(12):
        if crops[i] is None:
            crops[i] = grid[i]
        else:
            det_count += 1

    return crops, det_count

def extract_path(prob_map):
    # argmax + light smoothing
    path = np.argmax(prob_map, axis=0).astype(np.float32)
    if len(path) >= 7:
        path = savgol_filter(path, 7, 2)
    return prob_map.shape[0] - path

# =========================
# 5) COMPUTE PATIENT (CACHED)
# =========================
patient_cache = OrderedDict()

def compute_patient(pid, tlen):
    out = np.zeros((12, tlen), dtype=np.float32)

    path = get_image_path(pid)
    if not path:
        return out

    img = safe_read_rgb(path)
    if img is None:
        return out

    fs = DEFAULT_FS

    crops, det_count = get_crops(img)

    # prepare batch
    tensors, scales, clean_crops = [], [], []
    for c in crops:
        c = preprocess_remove_red_grid(c)
        clean_crops.append(c)
        h, w = c.shape[:2]
        s = TARGET_H / max(h, 1)
        nw = int(w * s)
        nw = min(nw, MAX_W)
        rz = cv2.resize(c, (nw, TARGET_H))
        t = torch.from_numpy(rz).permute(2,0,1).float()/255.0
        tensors.append(t)
        scales.append(s)

    mw = max([t.shape[2] for t in tensors])
    mw = int(np.ceil(mw/32)*32)
    batch = torch.zeros((12,3,TARGET_H,mw), device=DEVICE)
    for i,t in enumerate(tensors):
        batch[i,:,:, :t.shape[2]] = t.to(DEVICE)

    with torch.no_grad():
        # TTA: forward + horizontal flip
        p1 = torch.sigmoid(unet_model(batch))
        p2 = torch.sigmoid(unet_model(torch.flip(batch, [3])))
        preds = (p1 + torch.flip(p2, [3])) / 2.0
        preds = preds.cpu().numpy()

    global_grid = robust_grid_spacing_px(img)
    if global_grid is None:
        global_grid = 25.0

    for i in range(12):
        w_orig = tensors[i].shape[2]
        prob = preds[i,0,:, :w_orig]

        raw = extract_path(prob)

        local_grid = robust_grid_spacing_px(clean_crops[i])
        grid_px = local_grid if (local_grid is not None) else global_grid

        # sanity for grid
        if not (8.0 <= grid_px <= 200.0):
            grid_px = 25.0

        # assume big box 5mm = 0.5mV => pixels per mV = grid*2
        ppmv = grid_px * 2.0 * scales[i]
        if ppmv < 10:
            ppmv = 250.0

        sig_mv = (raw - np.median(raw)) / ppmv
        sig_mv = apply_filters(sig_mv, fs)
        sig_mv = sig_mv - np.median(sig_mv)

        # Light sniper per-lead (zeros only for this lead)
        if not lead_quality_ok(sig_mv):
            continue

        out[i] = resample(sig_mv, tlen)

    # mild Einthoven blend (safe)
    if np.any(out[1]):
        out[1] = 0.5*out[1] + 0.5*(out[0] + out[2])

    return out

# =========================
# 6) WRITE SUBMISSION
# =========================
print("🚀 Writing submission.csv (V46 BALANCED)...")
with open(SUBMISSION_FILE, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["id","value"])

    for rid in tqdm(template_ids, desc="Processing"):
        try:
            pid, idx, lead = rid.rsplit("_", 2)
            idx = int(idx)
            tlen = patient_len.get(pid, 5000)

            if pid in patient_cache:
                mat = patient_cache[pid]
                patient_cache.move_to_end(pid)
            else:
                mat = compute_patient(pid, tlen)
                patient_cache[pid] = mat
                if len(patient_cache) > MAX_CACHE:
                    patient_cache.popitem(last=False)

            li = LEAD_TO_IDX.get(lead, 0)
            val = 0.0
            if 0 <= idx < tlen:
                v = float(mat[li][idx])
                if np.isfinite(v):
                    val = v
            w.writerow([rid, f"{val:.4f}"])
        except:
            w.writerow([rid, "0.0000"])

del patient_cache
gc.collect()
torch.cuda.empty_cache()
print("✅ Done. Ready to Submit.")
```
