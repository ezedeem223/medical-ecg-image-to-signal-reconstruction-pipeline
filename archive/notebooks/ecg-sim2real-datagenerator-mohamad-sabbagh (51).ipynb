```python
# --- Cell 22: V39+++ Ultimate (Robust ID Parser + NEW EffB3 UNet + Safe Calibration + Optional OLD fallback) ---

import gc, os, csv, glob, math
import numpy as np
import pandas as pd
import cv2
import torch
from collections import OrderedDict
from tqdm import tqdm
from scipy.signal import savgol_filter, resample, find_peaks, butter, filtfilt
import segmentation_models_pytorch as smp
from ultralytics import YOLO

# ---------------------------
# 0) Constants
# ---------------------------
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

TEST_CSV_PATH = "/kaggle/input/physionet-ecg-image-digitization/test.csv"
SAMPLE_PARQUET_PATH = "/kaggle/input/physionet-ecg-image-digitization/sample_submission.parquet"
SUBMISSION_FILE = "submission.csv"
IMAGE_DIR = "/kaggle/input/physionet-ecg-image-digitization"

# These are defined in Cell 2 (resolved robustly)
try:
    NEW_UNET_PATH
    OLD_UNET_PATH
    YOLO_PATH
except NameError:
    NEW_UNET_PATH = None
    OLD_UNET_PATH = None
    YOLO_PATH = None

LEAD_NAMES = ['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']
LEAD_TO_IDX = {name: i for i, name in enumerate(LEAD_NAMES)}

TARGET_H = 256
YOLO_INF_MAX = 1280
YOLO_CONF = 0.10
MAX_W = 2048
MAX_CACHE = 12

DP_MAX_W = 1400
DP_SMOOTH = 0.45

# Old model fallback usage
USE_OLD_FALLBACK = True if (OLD_UNET_PATH is not None and os.path.exists(OLD_UNET_PATH)) else False

print(f"⚡ Device: {DEVICE}")
print("📌 Paths:")
print(" - NEW_UNET_PATH:", NEW_UNET_PATH, "| exists:", bool(NEW_UNET_PATH and os.path.exists(NEW_UNET_PATH)))
print(" - OLD_UNET_PATH:", OLD_UNET_PATH, "| exists:", bool(OLD_UNET_PATH and os.path.exists(OLD_UNET_PATH)))
print(" - YOLO_PATH    :", YOLO_PATH,     "| exists:", bool(YOLO_PATH and os.path.exists(YOLO_PATH)))
print("USE_OLD_FALLBACK:", USE_OLD_FALLBACK)

# ---------------------------
# 1) Read template ids
# ---------------------------
if not os.path.exists(SAMPLE_PARQUET_PATH):
    raise FileNotFoundError("❌ sample_submission.parquet not found")

print("📦 Reading Parquet template ids...")
template = pd.read_parquet(SAMPLE_PARQUET_PATH, columns=["id"])
template_ids = template["id"].astype(str).to_numpy()
del template
gc.collect()
print(f"✅ Template rows: {len(template_ids):,}")

# ---------------------------
# 2) Robust ID parser (FIXES: zeros / wrong parsing / lead mapping)
# ---------------------------
def parse_rid(rid: str):
    """
    Supports multiple possible id formats:
    - pid_idx_leadName
    - pid_leadName_idx
    - pid_idx_leadIndex(0-11)  (numeric)
    - pid_leadIndex_idx
    - pid may contain underscores
    Returns: (pid, idx:int, lead_idx:int) or None
    """
    s = str(rid).strip()
    parts = s.split("_")
    if len(parts) < 3:
        return None

    # helper: lead token to lead_idx
    def lead_token_to_idx(tok):
        tok = str(tok).strip()
        if tok in LEAD_TO_IDX:
            return LEAD_TO_IDX[tok]
        # numeric lead idx
        if tok.isdigit():
            li = int(tok)
            if 0 <= li < 12:
                return li
        return None

    # Case A: ... _ idx _ lead
    li = lead_token_to_idx(parts[-1])
    if li is not None and parts[-2].lstrip("-").isdigit():
        idx = int(parts[-2])
        pid = "_".join(parts[:-2])
        return pid, idx, li

    # Case B: ... _ lead _ idx
    li = lead_token_to_idx(parts[-2])
    if li is not None and parts[-1].lstrip("-").isdigit():
        idx = int(parts[-1])
        pid = "_".join(parts[:-2])
        return pid, idx, li

    # Case C: last two are digits => assume idx + leadIndex (or leadIndex + idx)
    if parts[-1].isdigit() and parts[-2].isdigit():
        a = int(parts[-2])
        b = int(parts[-1])
        # try interpret b as lead index
        if 0 <= b < 12:
            pid = "_".join(parts[:-2])
            return pid, a, b
        # try interpret a as lead index
        if 0 <= a < 12:
            pid = "_".join(parts[:-2])
            return pid, b, a

    return None

# Patient lengths map (based on idx max)
patient_lengths = {}
print("🧠 Building patient lengths from template ids (robust)...")
for rid in tqdm(template_ids, desc="Build Lengths"):
    p = parse_rid(rid)
    if p is None:
        continue
    pid, idx, _ = p
    cur = patient_lengths.get(pid, 0)
    if idx + 1 > cur:
        patient_lengths[pid] = idx + 1
print(f"✅ patient_lengths mapped for {len(patient_lengths):,} patients.")

# ---------------------------
# 3) Index images + fs_map (best-effort)
# ---------------------------
print("🗂️ Indexing images...")
image_paths = glob.glob(f"{IMAGE_DIR}/**/*.png", recursive=True) + glob.glob(f"{IMAGE_DIR}/**/*.jpg", recursive=True)
path_map = {os.path.splitext(os.path.basename(p))[0]: p for p in image_paths}
print(f"✅ Indexed images: {len(path_map):,}")

def clean_pid(pid):
    s = str(pid).strip()
    # remove trailing .0 if exists
    return s[:-2] if s.endswith(".0") else s

def get_image_path(pid):
    pid_s = clean_pid(pid)
    if pid_s in path_map:
        return path_map[pid_s]
    # try numeric normalization
    try:
        return path_map.get(str(int(float(pid_s))))
    except:
        return None

fs_map = {}
if os.path.exists(TEST_CSV_PATH):
    try:
        tdf = pd.read_csv(TEST_CSV_PATH, dtype=str)
        cols_low = {c.lower(): c for c in tdf.columns}
        id_c = next((cols_low[c] for c in cols_low if c == "id" or "id" in c), None)
        fs_c = next((cols_low[c] for c in cols_low if c == "fs" or "sampling" in c), None)
        if id_c and fs_c:
            fs_map = dict(zip(tdf[id_c].astype(str).map(clean_pid), tdf[fs_c].astype(str)))
        print(f"✅ fs_map loaded: {len(fs_map):,} items")
    except Exception as e:
        print(f"⚠️ fs_map read failed: {e}")

# ---------------------------
# 4) Load YOLO + Lead name mapping
# ---------------------------
print("⚙️ Loading YOLO (offline)...")
yolo_model = None
CLASS_TO_LEAD_IDX = {}

def _normalize_lead_name(x: str) -> str:
    s = str(x).strip()
    s = s.replace("Lead", "").replace("lead", "").replace("_", "").replace("-", "").replace(" ", "")
    s = s.replace("AVR", "aVR").replace("AVL", "aVL").replace("AVF", "aVF")
    s = s.replace("AVr","aVR").replace("AVl","aVL").replace("AVf","aVF")
    return s

if YOLO_PATH and os.path.exists(YOLO_PATH):
    try:
        yolo_model = YOLO(YOLO_PATH)
        print(f"✅ YOLO loaded: {YOLO_PATH}")

        names = getattr(yolo_model, "names", None)
        if isinstance(names, dict):
            items = list(names.items())
        elif isinstance(names, list):
            items = list(enumerate(names))
        else:
            items = []

        for cid, cname in items:
            n = _normalize_lead_name(cname)
            if n in LEAD_TO_IDX:
                CLASS_TO_LEAD_IDX[int(cid)] = LEAD_TO_IDX[n]

        # fallback identity if mapping incomplete
        if len(CLASS_TO_LEAD_IDX) < 8:
            print(f"⚠️ YOLO class mapping incomplete ({len(CLASS_TO_LEAD_IDX)}/12). Using identity fallback.")
            for i in range(12):
                CLASS_TO_LEAD_IDX.setdefault(i, i)

        print(f"✅ CLASS_TO_LEAD_IDX: {CLASS_TO_LEAD_IDX}")
    except Exception as e:
        print(f"⚠️ YOLO load failed: {e}")

# ---------------------------
# 5) Load UNet models (NEW EffB3 + optional OLD)
# ---------------------------
print("⚙️ Loading UNet models (offline)...")

# NEW: must match training architecture: Unet + EfficientNet-B3 + scSE
new_unet = smp.Unet(
    encoder_name="efficientnet-b3",
    encoder_weights=None,          # مهم: لا تحميل من النت
    in_channels=3,
    classes=1,
    decoder_attention_type="scse",
).to(DEVICE).eval()

def load_state_safely(model, path):
    sd = torch.load(path, map_location=DEVICE)
    # if it's EMA dict (missing non-float buffers), merge into model state
    if isinstance(sd, dict):
        msd = model.state_dict()
        msd.update(sd)
        model.load_state_dict(msd, strict=False)
    else:
        model.load_state_dict(sd, strict=False)
    return model

if NEW_UNET_PATH and os.path.exists(NEW_UNET_PATH):
    new_unet = load_state_safely(new_unet, NEW_UNET_PATH)
    print(f"✅ NEW UNet loaded: {NEW_UNET_PATH}")
else:
    raise FileNotFoundError("❌ NEW_UNET_PATH not found. Ensure best_model_effb3_phase9_ddp.pth is in your dataset input.")

old_unet = None
if USE_OLD_FALLBACK:
    # OLD model architecture (as your previous notebook): Unet + ResNet50 + scSE
    old_unet = smp.Unet(
        encoder_name="resnet50",
        encoder_weights=None,
        in_channels=3,
        classes=1,
        decoder_attention_type="scse",
    ).to(DEVICE).eval()

    try:
        old_unet = load_state_safely(old_unet, OLD_UNET_PATH)
        print(f"✅ OLD UNet loaded: {OLD_UNET_PATH}")
    except Exception as e:
        print(f"⚠️ OLD UNet load failed: {e}")
        old_unet = None
        USE_OLD_FALLBACK = False

# ---------------------------
# 6) Helpers
# ---------------------------
def safe_read_rgb(path):
    try:
        img = cv2.imread(path)
        if img is None:
            return None
        return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    except:
        return None

def preprocess_remove_grid_rgb(img_rgb):
    # remove red/pink grid lightly
    try:
        hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
        mask = (cv2.inRange(hsv, (0, 50, 50), (10, 255, 255)) |
                cv2.inRange(hsv, (170, 50, 50), (180, 255, 255)))
        out = img_rgb.copy()
        out[mask > 0] = (255, 255, 255)
        return out
    except:
        return img_rgb

def apply_high_pass(data, cutoff=0.5, fs=500.0, order=3):
    try:
        if len(data) < order * 3:
            return data
        nyq = 0.5 * float(fs)
        if nyq <= 0:
            return data
        wn = cutoff / nyq
        if not (0 < wn < 1):
            return data
        b, a = butter(order, wn, btype="high")
        return filtfilt(b, a, data)
    except:
        return data

def smart_einthoven_fix(leads):
    try:
        if 'I' in leads and 'II' in leads and 'III' in leads:
            l = min(len(leads['I']), len(leads['II']), len(leads['III']))
            I = leads['I'][:l]
            II = leads['II'][:l]
            III = leads['III'][:l]
            residual = (I + III) - II
            leads['II'][:l] += residual / 3.0
            leads['I'][:l]  -= residual / 3.0
            leads['III'][:l]-= residual / 3.0
    except:
        pass
    return leads

# ---------------------------
# 7) Robust calibration (conservative)
# ---------------------------
def estimate_grid_spacing_px(img_rgb):
    try:
        gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
        if gray.size == 0 or np.std(gray) < 3:
            return None

        ex = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
        ey = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
        sx = np.sum(np.abs(ex), axis=0)
        sy = np.sum(np.abs(ey), axis=1)

        px, _ = find_peaks(sx, distance=10, prominence=np.percentile(sx, 75))
        py, _ = find_peaks(sy, distance=10, prominence=np.percentile(sy, 75))

        diffs = []
        if len(px) > 5:
            dx = np.diff(px)
            diffs.extend(dx[(dx > 8) & (dx < 120)])
        if len(py) > 5:
            dy = np.diff(py)
            diffs.extend(dy[(dy > 8) & (dy < 120)])

        if len(diffs) < 6:
            return None

        g = float(np.median(np.array(diffs)))
        if not (6 <= g <= 120):
            return None
        return g
    except:
        return None

def choose_ppmv(local_grid_px, resize_scale, raw_trace_px):
    """
    Conservative ppmv:
    - If grid unreliable -> fallback constant
    - Try small-box vs big-box and pick plausible amplitude
    """
    FALLBACK_PPMV_RESIZED = 220.0

    if local_grid_px is None or (not np.isfinite(local_grid_px)) or local_grid_px < 6 or local_grid_px > 120:
        return FALLBACK_PPMV_RESIZED

    ppmv_A = (local_grid_px * 10.0) * resize_scale  # treat as 1mm
    ppmv_B = (local_grid_px * 2.0)  * resize_scale  # treat as 5mm

    def score(ppmv):
        if ppmv <= 0:
            return 1e9
        sig = (raw_trace_px - np.median(raw_trace_px)) / ppmv
        sig = np.nan_to_num(sig)
        p2p = float(np.percentile(sig, 99) - np.percentile(sig, 1))
        if (not np.isfinite(p2p)) or p2p <= 0:
            return 1e9
        if p2p < 0.3:
            return 1000 + (0.3 - p2p) * 1000
        if p2p > 6.0:
            return 1000 + (p2p - 6.0) * 200
        return abs(p2p - 2.0)

    sA = score(ppmv_A)
    sB = score(ppmv_B)
    ppmv = ppmv_A if sA <= sB else ppmv_B

    if not np.isfinite(ppmv) or ppmv < 30 or ppmv > 800:
        return FALLBACK_PPMV_RESIZED
    return float(ppmv)

def amplitude_safety(sig_mv):
    """
    SAFE MODE:
    - No rescale (rescale can destroy SNR)
    - Only sanitize + clamp
    """
    sig_mv = np.nan_to_num(sig_mv, nan=0.0, posinf=0.0, neginf=0.0)
    return np.clip(sig_mv, -5.0, 5.0)

# ---------------------------
# 8) Adaptive Viterbi (DP / fast)
# ---------------------------
def viterbi_dp(prob_map, smooth=0.45):
    H, W = prob_map.shape
    cost = -np.log(prob_map + 1e-6).astype(np.float32)
    dp = np.zeros_like(cost, dtype=np.float32)
    parent = np.zeros((H, W), dtype=np.int16)

    dp[:, 0] = cost[:, 0]
    parent[:, 0] = np.arange(H, dtype=np.int16)

    for t in range(1, W):
        prev = dp[:, t-1]
        c_m1 = np.roll(prev, 1);  c_m1[0] = 1e9
        c_0  = prev
        c_p1 = np.roll(prev, -1); c_p1[-1] = 1e9

        stacked = np.stack([c_m1 + smooth, c_0, c_p1 + smooth], axis=0)
        which = np.argmin(stacked, axis=0).astype(np.int16)
        best_prev = stacked[which, np.arange(H)]

        dp[:, t] = cost[:, t] + best_prev
        parent[:, t] = np.clip(np.arange(H, dtype=np.int16) + (which - 1), 0, H-1)

    path = np.zeros(W, dtype=np.int32)
    path[-1] = int(np.argmin(dp[:, -1]))
    for t in range(W-2, -1, -1):
        path[t] = int(parent[path[t+1], t+1])

    return (H - path).astype(np.float32)

def viterbi_adaptive(prob_map):
    H, W = prob_map.shape
    if W <= 20 or H <= 5:
        path = np.argmax(prob_map, axis=0)
        return (H - path).astype(np.float32)

    if W <= DP_MAX_W:
        return viterbi_dp(prob_map, smooth=DP_SMOOTH)

    path = np.argmax(prob_map, axis=0).astype(np.float32)
    win = 21 if W >= 21 else (W if W % 2 == 1 else W-1)
    if win >= 5:
        path = savgol_filter(path, win, 2)
    return (H - path).astype(np.float32)

# ---------------------------
# 9) YOLO crops (mapped)
# ---------------------------
def get_crops_yolo_mapped(img_rgb):
    crops = [None] * 12
    if yolo_model is None:
        return crops

    h0, w0 = img_rgb.shape[:2]
    if h0 < 20 or w0 < 20:
        return crops

    scale = YOLO_INF_MAX / float(max(h0, w0))
    if scale < 1.0:
        w1, h1 = max(32, int(w0 * scale)), max(32, int(h0 * scale))
        img_inf = cv2.resize(img_rgb, (w1, h1))
    else:
        img_inf = img_rgb
        scale = 1.0

    best = {}
    try:
        res = yolo_model.predict(img_inf, conf=YOLO_CONF, verbose=False, device=DEVICE)
        if res and res[0].boxes is not None and len(res[0].boxes) > 0:
            boxes = res[0].boxes.data.detach().cpu().numpy()
            for b in boxes:
                x1, y1, x2, y2, conf, cls_id = b[:6]
                cls_id = int(cls_id)
                conf = float(conf)

                target_idx = CLASS_TO_LEAD_IDX.get(cls_id, cls_id)
                if not (0 <= target_idx < 12):
                    continue

                x1o, y1o, x2o, y2o = x1 / scale, y1 / scale, x2 / scale, y2 / scale
                x1o, y1o = max(0, int(x1o)), max(0, int(y1o))
                x2o, y2o = min(w0, int(x2o)), min(h0, int(y2o))

                if x2o <= x1o + 10 or y2o <= y1o + 10:
                    continue

                prev = best.get(target_idx)
                if (prev is None) or (conf > prev[0]):
                    best[target_idx] = (conf, (x1o, y1o, x2o, y2o))

        for li, (_, (x1o, y1o, x2o, y2o)) in best.items():
            crops[li] = img_rgb[y1o:y2o, x1o:x2o]

        return crops
    except:
        return crops

# ---------------------------
# 10) UNet batch predict (NEW first, OLD fallback only if needed)
# ---------------------------
def prepare_unet_batch(crops_rgb):
    tens_list, scales, widths = [], [], []
    for c in crops_rgb:
        h, w = c.shape[:2]
        if h < 5 or w < 5:
            continue
        sc = TARGET_H / float(h)
        nw = int(max(1, w * sc))
        if nw > MAX_W:
            nw = MAX_W
            sc = nw / float(w)
        rz = cv2.resize(c, (nw, TARGET_H))
        t = torch.from_numpy(rz).permute(2, 0, 1).float() / 255.0
        tens_list.append(t)
        scales.append(sc)
        widths.append(nw)

    if not tens_list:
        return None, None, None

    max_w = int(np.ceil(max(widths) / 32.0) * 32)
    max_w = min(max_w, MAX_W)

    batch = torch.zeros((len(tens_list), 3, TARGET_H, max_w), dtype=torch.float32, device=DEVICE)
    for i, t in enumerate(tens_list):
        w_i = min(t.shape[2], max_w)
        batch[i, :, :, :w_i] = t[:, :, :w_i].to(DEVICE)
    return batch, scales, widths

@torch.no_grad()
def model_predict_prob(model, batch, widths):
    if DEVICE == "cuda":
        with torch.cuda.amp.autocast(enabled=True):
            p1 = torch.sigmoid(model(batch))
            p2 = torch.sigmoid(model(torch.flip(batch, [3])))
            preds = (p1 + torch.flip(p2, [3])) / 2.0
    else:
        p1 = torch.sigmoid(model(batch))
        p2 = torch.sigmoid(model(torch.flip(batch, [3])))
        preds = (p1 + torch.flip(p2, [3])) / 2.0

    preds = preds.detach().float().cpu().numpy()
    prob_maps = []
    q95s = []
    for i in range(preds.shape[0]):
        w_i = widths[i]
        pm = preds[i, 0, :, :w_i].astype(np.float32)
        prob_maps.append(pm)
        q95s.append(float(np.percentile(pm, 95)))
    return prob_maps, q95s

def predict_unet_probmaps(crops_rgb):
    batch, scales, widths = prepare_unet_batch(crops_rgb)
    if batch is None:
        return [], []

    try:
        prob_new, q95_new = model_predict_prob(new_unet, batch, widths)

        # fallback old only for "bad confidence" cases
        if USE_OLD_FALLBACK and old_unet is not None:
            # threshold conservative: if 95th percentile too low => model maybe failed
            need_old = [i for i, q in enumerate(q95_new) if q < 0.25]
            if need_old:
                prob_old, _ = model_predict_prob(old_unet, batch, widths)
                for i in need_old:
                    # blend (old helps when new misses trace)
                    prob_new[i] = 0.6 * prob_new[i] + 0.4 * prob_old[i]

        return prob_new, scales

    except RuntimeError as e:
        if "out of memory" in str(e).lower():
            torch.cuda.empty_cache()
            return [], []
        raise e

# ---------------------------
# 11) Patient compute
# ---------------------------
patient_cache = OrderedDict()

def compute_patient_leads(pid, target_len):
    out = np.zeros((12, target_len), dtype=np.float32)

    path = get_image_path(pid)
    if not path:
        return out
    img = safe_read_rgb(path)
    if img is None:
        return out

    fs = float(fs_map.get(clean_pid(pid), 500.0))
    if fs <= 0:
        fs = 500.0

    crops = get_crops_yolo_mapped(img)
    detected = [(i, c) for i, c in enumerate(crops) if c is not None and c.size > 200]
    if not detected:
        return out

    lead_indices = [i for i, _ in detected]
    crops_rgb = [preprocess_remove_grid_rgb(c) for _, c in detected]

    prob_maps, scales = predict_unet_probmaps(crops_rgb)
    if not prob_maps:
        return out

    global_grid = estimate_grid_spacing_px(img)

    leads_dict = {}
    for j in range(len(prob_maps)):
        lead_idx = lead_indices[j]
        prob = prob_maps[j]
        sc = scales[j]

        raw_px = viterbi_adaptive(prob)

        local_grid = estimate_grid_spacing_px(detected[j][1])
        if local_grid is None:
            local_grid = global_grid

        ppmv = choose_ppmv(local_grid, sc, raw_px)
        sig_mv = (raw_px - np.median(raw_px)) / ppmv
        sig_mv = np.nan_to_num(sig_mv, nan=0.0, posinf=0.0, neginf=0.0)

        if len(sig_mv) >= 11:
            sig_mv = savgol_filter(sig_mv, 11, 3)
        if len(sig_mv) >= 30:
            sig_mv = apply_high_pass(sig_mv, cutoff=0.5, fs=fs, order=3)

        # sanity fallback (if amplitude absurd)
        p2p = float(np.percentile(sig_mv, 99) - np.percentile(sig_mv, 1)) if len(sig_mv) > 10 else 0.0
        if (not np.isfinite(p2p)) or (p2p < 0.05) or (p2p > 12.0):
            sig_mv = (raw_px - np.median(raw_px)) / 220.0
            if len(sig_mv) >= 11:
                sig_mv = savgol_filter(sig_mv, 11, 3)
            if len(sig_mv) >= 30:
                sig_mv = apply_high_pass(sig_mv, cutoff=0.5, fs=fs, order=3)

        sig_mv = amplitude_safety(sig_mv)

        if len(sig_mv) > 0:
            leads_dict[LEAD_NAMES[lead_idx]] = resample(sig_mv, target_len).astype(np.float32)

    leads_dict = smart_einthoven_fix(leads_dict)

    for i, name in enumerate(LEAD_NAMES):
        if name in leads_dict:
            out[i] = np.nan_to_num(leads_dict[name], nan=0.0)

    return out

# ---------------------------
# 12) Streaming write (robust parse)
# ---------------------------
print("🚀 Writing submission.csv (V39+++ Ultimate)...")

bad_parse = 0
with open(SUBMISSION_FILE, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["id", "value"])

    for rid in tqdm(template_ids, desc="Processing Rows"):
        p = parse_rid(rid)
        if p is None:
            bad_parse += 1
            writer.writerow([rid, "0.0000"])
            continue

        pid_raw, idx, lead_idx = p
        pid = clean_pid(pid_raw)

        t_len = int(patient_lengths.get(pid_raw, patient_lengths.get(pid, 0)) or 0)
        # if length unknown, infer from fs
        if t_len <= 0:
            fs = float(fs_map.get(clean_pid(pid), 500.0))
            t_len = 5000 if fs >= 400 else 2500

        if t_len <= 0 or t_len > 20000:
            t_len = 5000

        # cache
        if pid in patient_cache:
            mtx = patient_cache[pid]
            patient_cache.move_to_end(pid)
        else:
            mtx = compute_patient_leads(pid, t_len)
            patient_cache[pid] = mtx
            if len(patient_cache) > MAX_CACHE:
                patient_cache.popitem(last=False)

        val = 0.0
        if 0 <= idx < mtx.shape[1] and 0 <= lead_idx < 12:
            v = float(mtx[lead_idx][idx])
            if np.isfinite(v):
                val = v

        writer.writerow([rid, f"{val:.4f}"])

print("✅ Done. submission.csv ready.")
print("Bad parse rows:", bad_parse, "out of", len(template_ids))

del patient_cache
gc.collect()
if torch.cuda.is_available():
    torch.cuda.empty_cache()
```
